## Step 1 - Import Data from Text File

<p>Copy and paste the raw text information into a variable called <code>data</code></p>

```python
import re

# Define the file path
file_path = "dummy_data.txt"

# Initialize an empty string to store the data
data = ""

# Read data from the file
with open(file_path, 'r') as file:
    data = file.read()

# Now you have the file content stored in the 'data' variable
print(data)  # This will display the content with newline characters intact
```

## Step 2 - Extract epoch

```python
# Initialize an empty list to store val_accuracy values
val_epoch = []

# Use regular expressions to find val_accuracy values
pattern = r'Epoch (\d+)/\d+'
matches = re.findall(pattern, data)

# Convert the matched strings to float and append to the list
for match in matches:
    val_epoch.append(float(match))
```

## Step 3 - Extract loss (training and testing)

```python
# Initialize an empty list to store val_accuracy values
val_loss_train = []

# Use regular expressions to find val_accuracy values
pattern = r'- loss: (\d+\.\d+)'
matches = re.findall(pattern, data)

# Convert the matched strings to float and append to the list
for match in matches:
    val_loss_train.append(float(match))

# Initialize an empty list to store val_accuracy values
val_loss_test = []

# Use regular expressions to find val_accuracy values
pattern = r'- val_loss: (\d+\.\d+)'
matches = re.findall(pattern, data)

# Convert the matched strings to float and append to the list
for match in matches:
    val_loss_test.append(float(match))
```

## Step 4 - Extract accuracy (training and testing)

```python
# Initialize an empty list to store val_accuracy values
val_accuracy_train = []

# Use regular expressions to find val_accuracy values
pattern = r'- accuracy: (\d+\.\d+)'
matches = re.findall(pattern, data)

# Convert the matched strings to float and append to the list
for match in matches:
    val_accuracy_train.append(float(match))

# Initialize an empty list to store val_accuracy values
val_accuracy_test = []

# Use regular expressions to find val_accuracy values
pattern = r'- val_accuracy: (\d+\.\d+)'
matches = re.findall(pattern, data)

# Convert the matched strings to float and append to the list
for match in matches:
    val_accuracy_test.append(float(match))
```

## Step 5 - Convert list to 1D matrix

```python
# Convert the list to a 1D matrix (NumPy array)
import numpy as np

epoch = np.array(val_epoch)
loss_train = np.array(val_loss_train)
loss_test = np.array(val_loss_test)
accuracy_train = np.array(val_accuracy_train)
accuracy_test = np.array(val_accuracy_test)
```

```python
print(len(loss_train))
print(len(loss_test))
print(len(epoch))
```

## Plotting Graph Loss (Training and Testing) vs Epoch

```python
import matplotlib.pyplot as plt

data_x = epoch
data_y1 = loss_train
data_y2 = loss_test

# Create a continuous line plot with a thinner line
plt.figure(figsize=(8, 6))
plt.plot(data_x, data_y1, label='Loss (Training)', marker='o', markersize=2, linestyle='-', color='blue', linewidth=1)
plt.plot(data_x, data_y2, label='Loss (Testing)', marker='o', markersize=2, linestyle='-', color='red', linewidth=1)

plt.title('Loss (Training and Testing) vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Save the graph as a .jpg image
plt.savefig('Loss (Training and Testing) vs. Epoch.jpg')

# Show the graph
plt.show()
```

## Plotting Graph Accuracy (Training and Testing) vs Epoch

```python
import matplotlib.pyplot as plt

data_x = epoch
data_y1 = accuracy_train
data_y2 = accuracy_test

# Create a continuous line plot with a thinner line
plt.figure(figsize=(8, 6))
plt.plot(data_x, data_y1, label='Accuracy (Training)', marker='o', markersize=2, linestyle='-', color='blue', linewidth=1)
plt.plot(data_x, data_y2, label='Accuracy (Testing)', marker='o', markersize=2, linestyle='-', color='red', linewidth=1)

plt.title('Accuracy (Training and Testing) vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()

# Save the graph as a .jpg image
plt.savefig('Accuracy (Training and Testing) vs. Epoch.jpg')

# Show the graph
plt.show()
```
